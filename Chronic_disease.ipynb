```python
def convert_to_numeric(df,column):
    df[column]=pd.to_numeric(df[column],errors='coerce')

for column in text_columns:
    convert_to_numeric(df,column)
    print(f"{column}:{df[column].dtype}")
```

```python
missing=df.isnull().sum()
missing[missing>0].sort_values(ascending=True,inplace=True)

missing_df=pd.DataFrame({
    'Missing Value':missing ,
    'Data Type':df.dtypes[missing.index]
})
missing_df
```

```python
def mean_values_imp(df,column):
    mean_value=df[column].mean()
    df[column].fillna(value=mean_value,inplace=True)
    
def mode_values_imp(df,column):
    mode_value=df[column].mode()[0]
    df[column].fillna(mode_value,inplace=True)
    
```

```python
num_cols=[col for col in df.columns if df[col].dtype!='object']

for col_name in num_cols :
   mean_values_imp(df,col_name) 
```

```python
cat_cols=[col for col in df.columns if df[col].dtype=='object']

for col_name in cat_cols:
    mode_values_imp(df,col_name)
```

```python
df['diabetes_mellitus']=df['diabetes_mellitus'].replace(to_replace={' yes':'yes','\tno':'no','\tyes':'yes'})
df['coranary_artery_disease']=df['coranary_artery_disease'].replace(to_replace={'\tno':'no'})
df['class']=df['class'].replace(to_replace={'ckd\t':'ckd','notckd':'not ckd'})
```

```python
df.head()
df['diabetes_mellitus'].unique()
```

```python
df['class']=df['class'].map({'ckd':1,'not ckd':2})
df['coranary_artery_disease']=df['coranary_artery_disease'].replace({'no':0,'yes':1})
df['red_blood_cells']=df['red_blood_cells'].map({'normal':1,'abnormal':0})
df['puscells']=df['puscells'].map({'normal':1,'abnormal':0})
df['pus_cell_clumps']=df['pus_cell_clumps'].map({'present':1,'notpresent':0})
df['bacteria']=df['bacteria'].map({'present':1,'notpresent':0})
df['hypertension']=df['hypertension'].map({'no':0,'yes':1})
df['diabetes_mellitus']=df['diabetes_mellitus'].map({'no':0,'yes':1})
df['appetite']=df['appetite'].map({'good':1,'poor':0})
df['peda_edema']=df['peda_edema'].map({'no':0,'yes':1})
df['anemia']=df['anemia'].map({'no':0,'yes':1})
```

```python
from sklearn.model_selection import train_test_split
```

```python
x=df.drop('class',axis=1)
y=df['class']

x_train,x_test,y_train,y_test=train_test_split(x,y,test_size=0.25,random_state=25)

print(f'X shape:{x_train.shape}')
print(f'x test shape:{x_test.shape}')
```

```python
from sklearn.tree import DecisionTreeClassifier

dct=DecisionTreeClassifier()
dct.fit(x_train,y_train)
```

```python
y_pred_dct=dct.predict(x_test)
y_pred_dct
```

```python
from sklearn.naive_bayes import GaussianNB
from sklearn.neighbors import KNeighborsClassifier
from sklearn.ensemble import RandomForestClassifier
from sklearn.svm import SVC
```

```python
models=[]
models.append(('Navie Bayes',GaussianNB()))
models.append(('KNN',KNeighborsClassifier(n_neighbors=8)))
models.append(('Random Forest Classifier',RandomForestClassifier()))
models.append(('Decision Tree classifier',DecisionTreeClassifier()))
models.append(('SVC',SVC(kernel='linear')))
```

```python
from sklearn.metrics import accuracy_score, precision_score, recall_score, f1_score, confusion_matrix
```

```python
for name,model in models:
    print(name,model)
    print()
    model.fit(x_train,y_train)
    y_pred=model.predict(x_test)
    print(confusion_matrix(y_test,y_pred))
    print('\n')
    print("Accuray:",accuracy_score(y_test,y_pred))
    print('\n')
    print("Precision:",precision_score(y_test,y_pred))
    print('\n')
    print("Recall",recall_score(y_test,y_pred))
    print('\n')
    print("Fiscore:",f1_score(y_test,y_pred))
    print('\n')
```
